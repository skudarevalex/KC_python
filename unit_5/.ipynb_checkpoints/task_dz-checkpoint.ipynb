```python
#Задание 15
data = df.copy()
data['dateAdded'] = pd.to_datetime(data['dateAdded'])
result = data.groupby(data['dateAdded'].dt.month)['id'].nunique().to_frame() # nunique() - выдает количество уникальных значений; to_frame() - обращает series в dataframe
result
```

```python
# Задание 16
data = df.copy()
data['dateAdded'] = pd.to_datetime(data['dateAdded'])
data['dateUpdated'] = pd.to_datetime(data['dateUpdated'])
update_delta = data['dateUpdated'] - data['dateAdded']
update_delta = update_delta.dt.days
data.insert (loc= 0 , column='update_delta', value=update_delta)
result = data.groupby(
    data['city']
).agg({
    'update_delta': 'mean',
    'latitude': 'max'
})
zep_mean = result.loc['Zephyrhills']['update_delta']
print(zep_mean)
result = result.reset_index()
result
```

```python
#Задание 18
data = df.copy()
data['dateAdded'] = pd.to_datetime(data['dateAdded'])
data['dateUpdated'] = pd.to_datetime(data['dateUpdated'])
update_delta = data['dateUpdated'] - data['dateAdded']
update_delta = update_delta.dt.days
data.insert (loc= 0 , column='update_delta', value=update_delta)
#data['update_delta'].describe()
data['update_delta'].median()
```

```python
#Задание 19
data = df.copy()
data_cat = data['categories'].dropna().str.split(pat=",").str.len()
data_l = data_cat.to_frame(name = 'lenght')
data_l = data_l[(data_l['lenght'] > 20)]
data_cat20 = data.loc[data_l.index]
result = data_cat20.groupby(
    data_cat20['province']
).agg({
    'longitude': 'min'
})
result = result.round(3).reset_index()
result.to_csv('result.csv', sep=';', index=False)
```
