```python
#Задание 12 создание pivot_table
playstore = df.copy()
unique_playstore = playstore.drop_duplicates(subset='App')
p1 = unique_playstore.rename(columns = lambda x: str.lower(x).replace(' ', '_'))
p1['price'] = p1['price'].str.replace('$', '').astype('float')
p2 = p1.rename(columns={"price": "mean_price", "rating": "mean_rating", "reviews": "mean_reviews"})
table = pd.pivot_table(p2, values=['mean_price', 'mean_rating', 'mean_reviews'], index=['category', 'type'],
                       aggfunc={'mean_price': "mean", 'mean_rating': "mean",'mean_reviews': "mean"})
```

```python
table['mean_price'] = table['mean_price'].round(2)
table['mean_rating'] = table['mean_rating'].round(1)
table['mean_reviews'] = table['mean_reviews'].round(2)
table.to_csv("final_table.csv")
table
```
